## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# Determine .95 and .75 quartile of Customers
percentile_95 = float(orders.groupby('CustomerID').agg({'amount_spent':'sum'}).quantile(.95))
percentile_75 = float(orders.groupby('CustomerID').agg({'amount_spent':'sum'}).quantile(.75))
```

```python
customers_spend = orders.groupby('CustomerID').agg({'amount_spent':'sum'})

#identify VIP customers
VIP_customers = list(customers_spend[customers_spend['amount_spent'] > percentile_95].index)

#identify Preferred customers
pref_customers = list(customers_spend[(customers_spend['amount_spent'] > percentile_75) & (customers_spend['amount_spent'] <= percentile_95)].index)

```

```python
#add column with Customer_Type

orders['Customer_Type'] = np.where(orders['CustomerID'].isin(VIP_customers), 'VIP', np.where(orders['CustomerID'].isin(pref_customers), 'Preferred',''))

```

# Q2: How to identify which country has the most VIP Customers?

# Q3: How to identify which country has the most VIP+Preferred Customers combined?

Provide your solution for Q2 below:

```python
# Count the VIPs in each Country
gb_VIP_by_country = orders[orders['Customer_Type'] == 'VIP'].groupby('Country').agg({'Customer_Type':'count'})

#Get Country with max number of VIPs
gb_VIP_by_country[gb_VIP_by_country['Customer_Type'] == gb_VIP_by_country['Customer_Type'].max()]

```

```python
# Count the VIPs in each Country
gb_VIPpref_by_country = orders[orders['Customer_Type'].isin(['VIP','Preferred'])].groupby('Country').agg({'Customer_Type':'count'})

#Get Country with max number of VIP or Preferred
gb_VIPpref_by_country[gb_VIPpref_by_country['Customer_Type'] == gb_VIPpref_by_country['Customer_Type'].max()]
```
